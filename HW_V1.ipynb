Кажется, что данные отсортированы по столбцу **Overall**, проверим

```python
def check(a):
    for i in range(len(a)-1):
        if a[i] < a[i+1]:
            return False
    return True

check(data.Overall.values)
```

Данные отсортированы по столбцу **Overall**

Заменяем все суммы на числа

```python
def tranform_sum(x):
    if type(x) == float:
        return x
    if x is None:
        return 0
    if x[-1] == 'K':
        return float(x[1:-1]) * 10 ** 3
    elif x[-1] == 'M':
        return float(x[1:-1]) * 10 ** 6
```

```python
for el in ['Value', 'Wage', 'Release Clause']:
    data[el] = data[el].map(tranform_sum)
```

Что же у нас с весом и ростом?

```python
def f_to_sm(x):
    x = x.split('\'')
    return int(x[0]) * 30.48 + int(x[1]) * 30.48 / 12
data['Height_sm'] = data['Height'].map(f_to_sm)
data['Weight_kg'] = data['Weight'].map(lambda x: int(x[:-3]) * 0.453592)
data['flag'] = 'football player'
```

```python
df_wh = pd.read_csv('weight-height.csv')
df_wh = df_wh[df_wh.Gender=='Male']
df_wh.Gender = df_wh.Gender.map({'Male': 'random man'})
df_wh['Height'] = df_wh['Height'] * 30.48 / 12
df_wh['Weight'] = df_wh['Weight'] * 0.453592

df_wh.columns = ['flag', 'Height_sm', 'Weight_kg']
```

```python
df = pd.concat([df_wh, data[['flag', 'Height_sm', 'Weight_kg']]])
```

```python
sns.relplot(x='Weight_kg', y='Height_sm', hue='flag', data=df)
```

Видно, что футболисты более атлетичны, чем случайные мужчины, также средняя зависимость между у ростом и весом у футболистов отличается от случайных мужчин

Сравним статистики средних игроков на разных позициях

```python
labels=['Crossing', 'Finishing', 'HeadingAccuracy', 'ShortPassing',
       'Volleys', 'Dribbling', 'Curve', 'FKAccuracy', 'LongPassing',
       'BallControl', 'Acceleration', 'SprintSpeed', 'Agility', 'Reactions',
       'Balance', 'ShotPower', 'Jumping', 'Stamina', 'Strength', 'LongShots',
       'Aggression', 'Interceptions', 'Positioning', 'Vision', 'Penalties',
       'Composure', 'Marking', 'StandingTackle', 'SlidingTackle', 'GKDiving',
       'GKHandling', 'GKKicking', 'GKPositioning', 'GKReflexes']
df = data.loc[:, ['Position'] + labels].groupby(by='Position').mean()
```

```python
position = ['GK', 'ST', 'CB']
stats = [[], [], []]
for i, el in enumerate(position):
    
    stats[i] = df.loc[el,:]
    angles = np.linspace(0, 2*np.pi, len(labels), endpoint=False)
    stats[i] = np.concatenate((stats[i],[stats[i][0]]))
    angles = np.concatenate((angles,[angles[0]]))
```

```python
fig = plt.figure(figsize=(20, 20))
ax = fig.add_subplot(111, polar=True)

ax.plot(angles, stats[0], 'ro-', linewidth=0.5, label=position[0])
ax.plot(angles, stats[1], 'bo-', linewidth=0.5, label=position[1])
ax.plot(angles, stats[2], 'go-', linewidth=0.5, label=position[2])
ax.fill(angles, stats[0], 'r', alpha=0.2)
ax.fill(angles, stats[1], 'b',alpha=0.2)
ax.fill(angles, stats[2], 'g',alpha=0.2)
ax.set_thetagrids(angles * 180/np.pi, labels)
ax.legend(fontsize=30)

ax.grid(True)

plt.show()
```
